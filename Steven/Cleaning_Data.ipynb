```python
import os 

import pandas as pd 

import numpy as py 
```

```python
#importing primary Saguaro Data Set 
full_data = "Full history of inventory and measurement data for all saguaros alive and dead_SR.csv"

```

```python
#converting file into a dataframe 
full_df = pd.read_csv(full_data,encoding='cp1252')

full_df
```

```python
#Using the group by method to isolate the right side of the data frame
#as_index is used so that the selected column headers aren't put into the dataframes index 
Measure_df  = full_df.groupby(['Accession Planting','Measurement Date','Height','Height Unit',
                               'DBH','DBH Unit','Measurement Note'], as_index=False).count()

```

```python
#Creating the data frame from the group by 
Measure_df = pd.DataFrame(Measure_df)

Measure_df
```

```python
#Dropping unwanted columns from the data frame
Measure_df = Measure_df.drop(columns=['Inventory Date',
                                      'Status','Condition','Inventory Note'
                       ,'DBH','DBH Unit'])
```

```python
#In order to isolate the year we are splitting the data the Measurement Date Column 
Measure_df[['Measurement Year','Time']] = Measure_df["Measurement Date"].str.split(" ",expand=True)

Measure_df
```

```python
Measure_df['Measurement Year'] = pd.to_datetime(Measure_df['Measurement Year'],errors='coerce').dt.year
```

```python
Measure_df['Measurement Year'].value_counts()
```

```python
#Converting the Height Column into a float variable 
Measure_df['Height'] = Measure_df['Height'].astype(float)

Measure_df
```

```python
#Converting inches to feet 
inch_value = Measure_df.loc[Measure_df['Height Unit']=="inches",'Height'] / 12

#replacing the inch metric with the values captured in the .loc function 
Measure_df.loc[(Measure_df['Height Unit']=='inches'),'Height'] == inch_value

Measure_df
```

```python
#renaming the columns Accession Planting and Height --> to the metric unit 
Measure_df.rename(columns={'Accession Planting':'Saguaro ID','Height':'Height in Feet'},inplace=True)

Measure_df
```

```python
#listing my column names to get proper calls 
list(Measure_df.columns)
```

```python
#dropping measurement date, time and height unit from columns 
Measure_df.drop(['Measurement Date', 'Time','Height Unit'], axis=1, inplace=True)
```

```python
#sorting the data by the measurement year 
Measure_df.sort_values(by='Measurement Year')
```

```python
#Creating a data frame based on the Inventory portion of the data set 
inventory_df = full_df.groupby(['Accession Planting','Inventory Date','Status','Condition','Inventory Note'], as_index=False).count()

inventory_df
```

```python
#Dropping unwanted columns from the data frame
inventory_df = inventory_df.drop(columns=['Height','Height Unit','Measurement Note',
                                      'Status','Measurement Date'
                       ,'DBH','DBH Unit'])
```

```python
inventory_df
```

```python
#In order to isolate the year we are splitting the data the Measurement Date Column 
inventory_df[['Measurement Year','Time']] = inventory_df["Inventory Date"].str.split(" ",expand=True)

inventory_df
```

```python
inventory_df['Measurement Year'] = pd.to_datetime(inventory_df['Measurement Year'],errors='coerce').dt.year
```

```python
inventory_df
```

```python
#renaming the columns Accession Planting and Height --> to the metric unit 
inventory_df.rename(columns={'Accession Planting':'Saguaro ID'},inplace=True)

inventory_df
```

```python
#dropping measurement date, time and height unit from columns 
inventory_df.drop(['Inventory Date', 'Time'], axis=1, inplace=True)

#sorting the data by the measurement year 
inventory_df.sort_values(by='Measurement Year')
```

```python
#Only keeping the rows in the data set that are greater than 2011 
inventory_df = inventory_df[inventory_df['Measurement Year'] > 2011]

inventory_df
```
